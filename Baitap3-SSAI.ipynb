```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv('filled_data.csv')

labeled_df = df[df['variety'].notnull()]
unlabeled_df = df[df['variety'].isnull()]

X_train = labeled_df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y_train = labeled_df['variety']
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

X_test = unlabeled_df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y_pred = knn.predict(X_test)
unlabeled_df['variety'] = y_pred

df = pd.concat([labeled_df, unlabeled_df])

df.to_csv('KNN.csv', index=False)
pd.set_option('display.max_rows', None)
df = pd.read_csv('KNN.csv')

print(df)
```

Ensemble learning

```python
import pandas as pd
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv('filled_data.csv')

labeled_df = df[df['variety'].notnull()]
unlabeled_df = df[df['variety'].isnull()]

X_train = labeled_df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y_train = labeled_df['variety']
bagging = BaggingClassifier(DecisionTreeClassifier(), n_estimators=100)
bagging.fit(X_train, y_train)

X_test = unlabeled_df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y_pred = bagging.predict(X_test)
unlabeled_df['variety'] = y_pred

df = pd.concat([labeled_df, unlabeled_df])

df.to_csv('EnsembleLearning.csv', index=False)
pd.set_option('display.max_rows', None)
df = pd.read_csv('LogisticRegression.csv')

print(df)
```
