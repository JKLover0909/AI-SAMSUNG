```python
#Câu 1 : vẽ biểu đồ cột cho neighborhood group
import pandas as pd
import matplotlib.pyplot as plt

file_path = "/content/AB_NYC_2019.csv"
data = pd.read_csv(file_path)

dulieu = data['neighbourhood_group']
cacgiatri = dulieu.value_counts()

plt.bar(cacgiatri.index, cacgiatri.values)

plt.xlabel('Khu vực')
plt.ylabel('Số lượng')
plt.title('Thống kê nhóm dân cư khu vực')
plt.show()
```

```python
#Câu 2 : tạo histogram cho neighbourhood
import pandas as pd
import matplotlib.pyplot as plt
file_path = "/content/AB_NYC_2019.csv"
data = pd.read_csv(file_path)

dulieu2 = data['neighbourhood']
cacgiatri2 = dulieu2.value_counts()

cacgiatri2

```

```python
plt.hist(cacgiatri2,bins = 8)
plt.xlabel('Số lần xuất hiện')
plt.ylabel('Tỉ lệ xuất hiện')
plt.title('Bảng đánh giá tỉ lệ xuất hiện')
plt.show
```

```python
#Câu 3 : Biểu diễn mối liên hệ giữa neighbourgroup và Availability của các phòng
import pandas as pd
import seaborn as sns

sns.boxplot(x='neighbourhood_group', y='availability_365', data=data)
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style('whitegrid')
plt.title('Biểu thị mối tương quan giữa Availability và Neighbor group')
plt.show
```

```python
#Câu 4 Vẽ bản đồ (scatter plot) của neighborhood dựa theo tọa độ lat lon
import pandas as pd
import seaborn as sns

sns.scatterplot(x='longitude', y='latitude', data=data)
sns.set(rc={'figure.figsize':(24,16)})
sns.set_style('white')
plt.title('Neighbourhood Map')
plt.show
```

```python
#Câu 5  Sử dụng heatmap để biểu diễn mối quan hệ (correlation) giữa tất cả các thuộc tính trong dữ liệu
import pandas as pd
import seaborn as sns
corr =  data.corr()
sns.heatmap(corr, annot = True)
#sns.set(rc={'figure.figsize':(4,4)})
#sns.set_style('whitegrid')
plt.title('Correlation Heatmap')
plt.show()
```
