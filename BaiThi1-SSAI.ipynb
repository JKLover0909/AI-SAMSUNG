# Tiền xử lí dữ liệu file Test_samsung_noclass.csv

```python
import pandas as pd

pd.set_option('display.max_rows', None)  # Hiển thị tất cả các hàng
pd.set_option('display.max_columns', None)  # Hiển thị tất cả các cột
pd.set_option('display.width', None)  # Hiển thị toàn bộ dữ liệu

def fill_missing_values(data):
    filled_data = data.copy()

    # Lặp qua từng cột trong DataFrame
    for column in filled_data.columns:
        # Nếu datatype của cột là object (không phải là số), điền các giá trị bị
        # thiếu cho trường rời rạc bằng phương pháp tần số cao nhất
        if filled_data[column].dtype == 'object':
            filled_data[column].fillna(filled_data[column].mode()[0], inplace=True)
        else:
            # Điền các giá trị bị thiếu cho trường số bằng phương pháp trung bình
            filled_data[column].fillna(round(filled_data[column].mean(), 2), inplace=True)
    return filled_data

data = pd.read_csv('Test_samsung_noclass.csv')

# Điền các giá trị bị thiếu
filled_data = fill_missing_values(data)

# Xuất dữ liệu đã điền ra file Test_samsung_noclass_imputed.csv
filled_data.to_csv('Test_samsung_noclass_imputed.csv', index=False)

#in dữ liệu đã
print(filled_data)
```

Mã hóa dữ liệu thành dummies variables

```python
import pandas as pd

# Đọc dữ liệu từ file CSV
data = pd.read_csv('Train_samsung_imputed.csv')

# Chọn các cột cần chuyển đổi
columns_to_encode = ['X1', 'X2', 'X3', 'X4', 'X5', 'X11']

# Áp dụng hàm get_dummies() để chuyển đổi dữ liệu thành dummies variables
encoded_data = pd.get_dummies(data, columns=columns_to_encode)

# Lưu kết quả vào file mới
encoded_data.to_csv('Train_Samsung_imputed_dummies.csv', index=False)
print(encoded_data)
```
